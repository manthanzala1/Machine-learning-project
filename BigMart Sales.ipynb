```python
# categorical to numerical transform
from sklearn.preprocessing import LabelEncoder
```

```python
df['newItem_Fat_Content'] = la.fit_transform(df['Item_Fat_Content'])
```

```python
df['newItem_Identifier'] = la.fit_transform(df['Item_Identifier'])
```

```python
df['newItem_Type'] = la.fit_transform(df['Item_Type'])
```

```python
df['newOutlet_Size'] = la.fit_transform(df['Outlet_Size'])
```

```python
df.drop(['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type'],axis = 1,inplace = True)
```

```python
x = df.drop('Item_Outlet_Sales',axis = 1).values
x
```

```python
y = df['Item_Outlet_Sales'].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pre = model.predict(x_test)
y_pre
```

```python
y_test
```

```python
model.score(x_test,y_test)
```
