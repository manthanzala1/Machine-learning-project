```python
# data visualizations
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
a = df['user_sex'].value_counts()
```

```python
plt.pie(x = a.values,labels = a.index,autopct = '%.2f%%')
plt.show()
```

```python
sns.histplot(df['user_age'],kde = True,bins = 10)
plt.show()
```

```python
zala = sns.countplot(df['user_country'])
zala.bar_label(container = zala.containers[0])
plt.xticks(rotation = 80)
plt.show()
```

```python
b = sns.barplot(x = "user_country",y = "rating",data = df)
b.bar_label(container = b.containers[0])
plt.xticks(rotation = 80)
plt.show()
```

```python
c = sns.countplot(df['favourite_character'])
c.bar_label(container = c.containers[0])
plt.xticks(rotation = 80)
plt.show()
```

```python
cor = df.corr()
sns.heatmap(cor,annot = True)
plt.show()
```
