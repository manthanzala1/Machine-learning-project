```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
la = LabelEncoder()
```

```python
df['newFraud_indicator'] = la.fit_transform(df['Fraud_indicator'])
```

```python
df
```

```python
x = df.drop(['Fraud_indicator','newFraud_indicator','Timestamp','FastagID','TollBoothID','Geographical_Location','Vehicle_Plate_Number'],axis = 1)
x
```

```python
y = df['newFraud_indicator']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
oh = OneHotEncoder()
```

```python
column_transformer = make_column_transformer((OneHotEncoder(sparse = False),['Vehicle_Type','Lane_Type','Vehicle_Dimensions']),remainder='passthrough')
```

```python
model = RandomForestClassifier(n_estimators=100,bootstrap=False)
```

```python
pipe = make_pipeline(column_transformer,model)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pre = pipe.predict(x_test)
```

```python
y_test
```

```python
pipe.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pre)
```
