```python
df.drop(['Applicant_ID','Marital_Status','House_Ownership','Vehicle_Ownership(car)'],axis = 1,inplace = True)
```

```python
df
```

```python
x = df.drop('Loan_Default_Risk',axis = 1)
x
```

```python
y = df['Loan_Default_Risk']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
column_transform = make_column_transformer((OneHotEncoder(sparse= False),['Occupation','Residence_City','Residence_State']),remainder='passthrough')
```

```python
from sklearn import tree
from sklearn.pipeline import make_pipeline
```

```python
model = tree.DecisionTreeClassifier()
```

```python
pipe = make_pipeline(column_transform,model)
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.score(x_test,y_test)
```

```python
y_pre = pipe.predict(x_test)
```

```python
y_pre
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pre)
```
